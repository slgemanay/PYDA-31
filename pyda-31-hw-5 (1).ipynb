```python
import json
i = 0
purchases = dict()
with open('purchase_log.txt', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        
        dict_ = json.loads(line)
        purchases[dict_['user_id']] = dict_['category']
```

```python
with open('visit_log.csv', 'r', encoding='utf-8') as f:
    with open('funnel.csv', 'w', encoding='utf-8') as f1:
        for line in f:
            id1 = dict_['user_id']
            if id1 in purchases:
                f1.write(line)
```
